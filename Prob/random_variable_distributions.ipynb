## 이항분포 분포

### 베르누이 확률 변수 실험

```python
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 사건이 1회인 경우
mu = 0.6
rv = sp.stats.bernoulli(mu)

xx = [0, 1]
plt.bar(xx, rv.pmf(xx))
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.ylabel('P(x)')
plt.show()
```

## 이항분포
베르누이 분포가 사건 1회라면, 이항분포는 이를 n번으로 확대한 경우다.

```python
N = 10  # 사건 횟수
mu = 0.6  # 성공 확률
rv = sp.stats.binom(N, mu)
```

```python
# 이항분포의 시각화
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align = "center")
plt.xlabel('samples')
plt.ylabel('P(x)')
plt.title('binomial distribution example')
plt.show()
```

## 정규분포

```python
# 정규분포의 정의
mu = 0
std = 1
rv = sp.stats.norm(mu, std)
```

```python
# 표준정규분포의 그래프
xx = np.linspace(-5, 5, 100)
plt.plot(xx, rv.pdf(xx))
plt.ylabel('P(X)')
plt.title('Standard Normal distribution')
plt.show()
```

```python
# 참고, 정규분포를 통한 랜덤 시드 얻기
np.random.seed(0)
x = rv.rvs(20)
x
```

## 포아송 분포

```python
from scipy.stats import poisson
x = np.arange(41, dtype = 'f8')
l = 20
y1 = poisson(l).pmf(x)
```

```python
plt.plot(x, y1, 'ro--')
plt.ylabel('P(X)')
plt.title('Poisson Distribution, lambda = 20')
plt.show()
```

### lambda 값에 따른 포아송 분포 그래프의 변화

```python
for l in range(3, 18, 3):
  plt.plot(x, poisson(l).pmf(x), 'o--', label='$(\lambda =$' + str(l) + ')')

plt.ylabel('P(X)')
plt.grid()
plt.legend()
plt.show()
```
