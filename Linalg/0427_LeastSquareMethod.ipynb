# 최소제곱법의 구현 (선형회귀를 통한 예시)

```python
import numpy as np
A = np.array([[-3, 1], [-1, 1], [1, 1], [3, 1]])
b = np.array([-1, -1, 3, 3])
```

```python
At = A.transpose()
```

```python
At_A = At @ A
print(At_A)
```

```python
At_b = At @ b
print(At_b)
```

```python
answer = np.linalg.inv(At_A) @ At_b
print(answer)
```

 즉, y = 0.8x + 1로 예측이 된다는 것이다. MSE를 확인해보자.

```python
b_new = A @ answer
print(b_new)
```

```python
sum([ x ** 2 for x in b - b_new ]) / 4
```
